```python
book_df= pd.read_csv(file,encoding='ISO-8859-1')
book_df.head()
```

```python
book=book_df.iloc[:,1:]
```

```python
book
```

```python
#rename the columns
book_cleaned= book.rename({'User.ID': 'userid','Book.Title':'title','Book.Rating':'rating'}, axis=1)
```

```python
#number of unique users in the dataset
len(book_cleaned.userid.unique())
```

```python
len(book_cleaned.title.unique())
```

```python
user_book_df = book_cleaned.pivot_table(index='userid',
                                 columns='title',
                                   values='rating').reset_index(drop=True)
```

```python
user_book_df
```

```python
user_book_df.index = book_cleaned.userid.unique()
```

```python
user_book_df
```

```python
#Impute those NaNs with 0 values
user_book_df.fillna(0, inplace=True)
```

```python
user_book_df
```

```python
user_book_df.tail()
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances( user_book_df.values,metric='cosine')
```

```python
user_sim
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
```

```python
#Set the index and column names to user ids 
user_sim_df.index = book_cleaned.userid.unique()
user_sim_df.columns = book_cleaned.userid.unique()
```

```python
user_sim_df.iloc[0:5, 0:5]
```

```python
#Most Similar Users
user_sim_df.idxmax(axis=1)[0:5]
```
