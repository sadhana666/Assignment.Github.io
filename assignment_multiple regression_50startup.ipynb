# Preparing a model

```python
#Build model
import statsmodels.formula.api as smf 
model = smf.ols('Profit~RandDSpend+Administration+MarketingSpend',data=startup1).fit()
```

```python
#Coefficients
model.params
```

```python
model.summary()
```

```python
#t and p-Values
print(model.tvalues, '\n', model.pvalues)
```

```python
#R squared values
(model.rsquared,model.rsquared_adj)
```

# Calculating VIF

```python
rsq_rd = smf.ols('RandDSpend~Administration+MarketingSpend',data=startup1).fit().rsquared  
vif_rd = 1/(1-rsq_rd) 

rsq_ad = smf.ols('Administration~RandDSpend+MarketingSpend',data=startup1).fit().rsquared  
vif_ad = 1/(1-rsq_ad) 

rsq_mk = smf.ols('MarketingSpend~Administration+RandDSpend',data=startup1).fit().rsquared  
vif_mk = 1/(1-rsq_mk) 


# Storing vif values in a data frame
d1 = {'Variables':['RandDSpend','Administration','MarketingSpend'],'VIF':[vif_rd,vif_ad,vif_mk]}
Vif_frame = pd.DataFrame(d1)  
Vif_frame
```

# Residual Analysis

```python
import statsmodels.api as sm
qqplot=sm.qqplot(model.resid,line='q') # line = 45 to draw the diagnoal line
plt.title("Normal Q-Q plot of residuals")
plt.show()
```

```python
list(np.where(model.resid>10))
```

# Residual Plot for Homoscedasticity

```python
def get_standardized_values( vals ):
    return (vals - vals.mean())/vals.std()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(get_standardized_values(model.fittedvalues),
            get_standardized_values(model.resid))

plt.title('Residual Plot')
plt.xlabel('Standardized Fitted values')
plt.ylabel('Standardized residual values')
plt.show()
```

# Residual Vs Regressors

```python
fig = plt.figure(figsize=(15,8))
fig = sm.graphics.plot_regress_exog(model, "RandDSpend", fig=fig)
plt.show()
```

```python
fig = plt.figure(figsize=(15,8))
fig = sm.graphics.plot_regress_exog(model, "Administration", fig=fig)
plt.show()
```

```python
fig = plt.figure(figsize=(15,8))
fig = sm.graphics.plot_regress_exog(model, "MarketingSpend", fig=fig)
plt.show()
```

# Cooks Distance

```python
model_influence = model.get_influence()
(c, _) = model_influence.cooks_distance
```

```python
#Plot the influencers values using stem plot
fig = plt.subplots(figsize=(20, 7))
plt.stem(np.arange(len(startup1)), np.round(c, 3))
plt.xlabel('Row index')
plt.ylabel('Cooks Distance')
plt.show()
```

```python
#index and value of influencer where c is more than .5
(np.argmax(c),np.max(c))
```
