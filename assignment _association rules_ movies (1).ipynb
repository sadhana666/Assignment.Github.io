```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
from mlxtend.frequent_patterns import association_rules,apriori
```

```python
data=pd.read_csv('C:/Users/LENOVO/Downloads/my_movies (2).csv')
data
```

```python
data1=data.iloc[:,5:]
data1
```

```python
frequent_set1=apriori(data1,min_support=0.05,max_len=2,use_colnames=True)
frequent_set1
```

```python
rule1=association_rules(frequent_set1,metric='lift',min_threshold=1)
rule1
```

```python
rule1.sort_values('lift',ascending = False)[0:20]
```

```python
rule1[rule1.lift>1]
```

```python
support1=pd.Series(rule1.support)
lift1=pd.Series(rule1.lift)
sns.scatterplot(x=support1,y=lift1)
```
